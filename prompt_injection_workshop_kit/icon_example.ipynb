```python
def insert_secret_text(visible_text: str, secret_text: str) -> str:
    """
    Fügt einem sichtbaren Text unsichtbaren Geheimtext hinzu.
    Menschen sehen nur den sichtbaren Text, Maschinen können den Geheimtext auslesen.
    """

    zero = "\u200b"  # Zero Width Space
    one = "\u200c"   # Zero Width Non-Joiner

    data = secret_text.encode("utf-8")
    bits = "".join(f"{byte:08b}" for byte in data)

    length_bits = f"{len(bits):032b}"
    hidden = "".join(one if bit == "1" else zero for bit in length_bits + bits)

    return visible_text + hidden


def decode_secret_text(text: str) -> str:
    """
    Liest den unsichtbaren Geheimtext aus einem Text aus.
    """

    zero = "\u200b"
    one = "\u200c"

    hidden_chars = [char for char in text if char in (zero, one)]

    if len(hidden_chars) < 32:
        return ""

    bits = "".join("1" if char == one else "0" for char in hidden_chars)

    length = int(bits[:32], 2)
    payload_bits = bits[32:32 + length]

    if len(payload_bits) != length or len(payload_bits) % 8 != 0:
        return ""

    data = bytes(
        int(payload_bits[i:i + 8], 2)
        for i in range(0, len(payload_bits), 8)
    )

    return data.decode("utf-8")
```

```python
text = "Hallo 😊 Wie geht es Ihnen?​​​​​​​​​​​​​​​​​​​​​​​​​‌‌‌‌​​​​‌‌​‌​​‌​‌‌​‌‌‌​​‌‌‌​‌​​​‌‌​​‌​‌​‌‌‌​​‌​​‌‌​‌‌‌​​‌‌​​‌​‌​​‌​‌‌​‌​‌‌​‌​​‌​‌‌​​‌​​​​‌​‌‌​‌​​‌‌​‌‌‌​​‌‌‌​​​​​‌‌​‌​​​​‌‌​​‌​"
secret = "interne-id-7842"

encoded = insert_secret_text(text, secret)

print(encoded)
# Für Menschen sichtbar:
# Hallo 😊 Wie geht es Ihnen?

decoded = decode_secret_text(encoded)

print(decoded)
# interne-id-7842
```
